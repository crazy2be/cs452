```python
%pylab inline
import csv
file = open('4/data.txt')
dataX = []
dataY = []
prevPos = (0, 0)
import math
#for (i, l) in enumerate(file):
#    l = l[len('Center: for 579:   '):]
#    pos = tuple([float(n) for n in l[1:-2].split(',')][:-1])
f = open('points.txt')
csvreader = csv.reader(f, delimiter=' ')
scale_factor = 1/60.0
for fields in csvreader:
    pos = (float(fields[1]), float(fields[2]))
    if math.isnan(pos[0]) or math.isnan(pos[1]):
        continue
    d = sqrt((pos[0]-prevPos[0])**2 + (pos[1]-prevPos[1])**2)
    prevPos = pos
    if d > 50: continue # Ignore the really fucked points
    dataX.append(float(fields[0]) * scale_factor) # Assume 30fps
    dataY.append(d / scale_factor)
   
dataX = np.asarray(dataX)
dataY = np.asarray(dataY)
dataYS = np.convolve(dataY, np.ones(20)/20)
dataYSi = [i for (i, d) in enumerate(dataY) if abs(d - dataYS[i]) < 5]
dataYSS = dataY[dataYSi]
dataXSS = dataX[dataYSi]
dataXSSS = dataXSS
dataYSSS = np.convolve(dataYSS, np.ones(20)/20)[10:-9]
```

```python
import scipy.optimize
import scipy.integrate
get_mean = lambda lst: sum(lst) / len(lst)

# fit a curve using non-linear optimization, with constraints

def fit_nlp(xs, ys):
    v0 = numpy.mean(ys[0:10])
    v1 = numpy.mean(ys[-10:])
    
    t0 = xs.min()
    t1 = xs.max() - t0

    def sqerr(coefs):
        return sum([(y - np.poly1d(coefs)(x - t0))**2 for x, y in zip(xs, ys)])

    def slope(coefs, x):
        n = coefs.shape[0]
        xs = np.power(np.full(n - 1, x), np.arange(n - 2, -1, -1)) * np.arange(n - 1, 0, -1)
        return np.dot(xs, coefs[:-1])

    constraints = ( # all constraints should equal 0
        #lambda coefs: np.poly1d(coefs)(0) - v0,
        #lambda coefs: np.poly1d(coefs)(t1) - v1,
        lambda coefs: slope(coefs, 0),
        lambda coefs: slope(coefs, t1),
    )

    guess = np.full(6, 0)

    solution = scipy.optimize.fmin_slsqp(sqerr, guess, constraints)

    return (solution, t0, t1)

solution, offset, t1_1 = fit_nlp(x1, y1)
model = np.poly1d(solution)

print(solution)
#print(list(map(lambda f: f(solution), constraints)))
#print(model(0))
#print(model(hi-lo))
#print(slope(solution, 0))
#print(slope(solution, hi - lo))

#print(guess)
#print(list(map(lambda f: f(guess), constraints)))

#figure(figsize=(10,3))
model_noclamp = numpy.poly1d(numpy.polyfit(x1 - offset, y1, 6))
fig = plt.figure(figsize=(10,3))
fig.suptitle("Deceleration model", fontsize=14)
plt.plot(x1 - offset, y1, 'k+')
plt.plot(x1 - offset, model(x1 - offset), 'b', label='Clamped')
plt.plot(x1 - offset, model_noclamp(x1 - offset), 'r', label='Unclamped')
plt.legend(fontsize=10)
plt.xlabel('Time (s)', fontsize=10)
plt.ylabel('Velocity (pixels/s)', fontsize=10)
fig.savefig('deceleration.png', dpi=200)
plt.show()
print(scipy.integrate.quad(model, 0, t1_1))
print(scipy.integrate.quad(model_noclamp, 0, t1_1))

def distance(xs, ys):
    return numpy.sum(ys[:-1] * (xs[1:] - xs[:-1]))

print(distance(x1, y1))

figure(figsize=(20,6))

solution2, offset2, t1_2 = fit_nlp(x2, y2)
model2 = np.poly1d(solution2)
scale = numpy.mean(y2[0:10]) / numpy.mean(y1[0:10])
print(scipy.integrate.quad(model2, 0, t1_2))
print(distance(x2, y2))
plt.plot(x2 - offset2, y2, 'b+', x2 - offset2, model2(x2 - offset2), 'k',  x2 - offset2, model((x2 - offset2) * 0.9) * scale, 'r')
```
